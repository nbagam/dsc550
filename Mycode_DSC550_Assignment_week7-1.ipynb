# 2. Drop the "Id" column and any features that are missing more than 40% of their values.

```python
house_price_data_df1= house_price_data.drop(['Id'],axis=1)
```

```python
house_price_data_df1
```

```python
missing_data_percent = (house_price_data_df1.isnull().sum()/len(house_price_data_df1)) * 100
df_columns_missing = pd.DataFrame({'column_name':house_price_data_df1.columns,
                                'missing_val_column':missing_data_percent})
```

```python
df_columns_missing= df_columns_missing.loc[df_columns_missing['missing_val_column'] >= 40]
```

```python
df_columns_missing
```

```python
#delete columns with more than 40% missing data
house_price_data_df1.drop(['PoolQC','MiscFeature','Alley','Fence','FireplaceQu'], axis = 1,inplace=True)
```

```python
house_price_data_df1.shape
```

# 3. For numerical columns, fill in any missing data with the median value.


```python
# Get numeric columns 
df_house_price_num = house_price_data_df1.select_dtypes(exclude='object')
```

```python
# Get Missing data numeric columns
numeric_missing_col = df_house_price_num.isnull().sum()
numeric_missing_col = numeric_missing_col[numeric_missing_col>0]
numeric_missing_col
```

```python
# Fill numeric missing data columns with median
house_price_data_df1[['LotFrontage','MasVnrArea','GarageYrBlt']] = house_price_data_df1[['LotFrontage','MasVnrArea','GarageYrBlt']].fillna('median')
```

```python
house_price_data_df1
```

# 4. For categorical columns, fill in any missing data with the most common value (mode).

```python
# Get categorical columns 
df_house_price_obj = house_price_data_df1.select_dtypes(include='object')
```

```python
df_house_price_obj
```

```python
obj_missing_col = df_house_price_obj.isnull().sum()
obj_missing_col = obj_missing_col[obj_missing_col>0]
obj_missing_col
```

```python
# Fill categorical missing data columns with mode value
house_price_data_df1[['MasVnrType', 'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1','BsmtFinType2', 'Electrical', 'GarageType', 'GarageFinish','GarageQual', 'GarageCond']] = house_price_data_df1[['MasVnrType', 'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1','BsmtFinType2', 'Electrical', 'GarageType', 'GarageFinish','GarageQual', 'GarageCond']].fillna(house_price_data.mode().iloc[0])
```

```python
house_price_data_df1
```

```python
# validate data for any null values.
house_price_data_df1.isnull().sum()
```

# 5.Convert the categorical columns to dummy variables.

```python
dum_house_data = pd.get_dummies(house_price_data_df1,drop_first=True)
```

```python
dum_house_data.head()
```

# 6.Split the data into a training and test set, where the SalePrice column is the target.

```python
x = dum_house_data.drop('SalePrice',axis=1)
y = dum_house_data['SalePrice']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20,random_state=101)
```

# 7.Run a linear regression and report the R2-value and RMSE on the test set.

```python
# Linear regression import

from sklearn.linear_model import LinearRegression
Lr = LinearRegression()
Lr.fit(x_train,y_train)
```

```python
y_pred = Lr.predict(x_test)
```

```python
# Report RMSE and R Square
from sklearn.metrics import mean_squared_error,r2_score
```

```python
MSE = mean_squared_error(y_test,y_pred).round(2)
RMSE = np.sqrt(MSE).round(2)
rsquare = Lr.score(x_train,y_train).round(2)
```
